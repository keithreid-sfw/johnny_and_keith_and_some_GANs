```julia
using Plots
using Test;
```

```julia
function get_layers()
    layers          = 5
    return layers
end

function get_wlength()
    wlength          = 1000
    return wlength
end;
```

```julia
function build_simple_wave(amplitude, freq_mod, shift)
    wlength     = get_wlength()
    x_steps     = 0:0.1:wlength
    simple_wave = [amplitude*sin((freq_mod*x)+shift) for x in x_steps]
    return simple_wave
end;
```

```julia
function build_complex_wave()
    
    length          = get_length()
    layers          = get_layers()
    complex_wave    = build_simple_wave(0,0,0)

    for i in 1:layers

        amplitude              = rand()
        freq_mod               = rand()
        shift                  = rand()
        
        this_simple_wave = build_simple_wave(amplitude, freq_mod, shift)
        complex_wave     = complex_wave+this_simple_wave
    end
    return complex_wave
end;
```

```julia
function wave_builder()
    complex_wave    = build_complex_wave()
    return complex_wave 
end

function main()
    complex_wave = wave_builder()
    complex_plot = plot(complex_wave, legend = :false)
    return complex_plot
end

complex_plot = main();
```

```julia
complex_plot
```

```julia
function test_build_complex_wave()
    complex_wave                    = build_complex_wave()
    @test typeof(complex_wave)      == Array{Float64, 1}
    @test typeof(complex_wave[1])   == Float64
end

function test_build_simple_wave()
    amplitude              = rand()
    freq_mod               = rand()
    shift                  = rand()
    this_simple_wave = build_simple_wave(amplitude, freq_mod, shift)
    @test typeof(this_simple_wave)      == Array{Float64, 1}
    @test typeof(this_simple_wave[1])   == Float64
end

function all_tests()
    test_build_complex_wave()
    test_build_simple_wave()
end

all_tests()
```
